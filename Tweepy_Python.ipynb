Opening tweets.txt to write and writing Twitter data in it. Collecting the exceptions and creating a list.

```python
exceptions_list = []

with open('tweets.txt', 'w', encoding = 'utf-8') as f:
    for id_tweet in tweet_list_id: 
        try:
            tweet = api.get_status(id_tweet, tweet_mode= 'extended')
            json.dump(tweet._json, f)
            f.write("\n")
        except Exception as e:
            exceptions_list.append(id_tweet)
            print(str(e))
```
